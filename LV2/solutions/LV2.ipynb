```python
#Zadatak 1
import re
list_emails=[]


fhand = open('mbox-short.txt') 
for line in fhand:
  line = line.rstrip()
  if(re.findall("\S+@\S+", line) != []):
    list_emails.append(re.findall("(\S+)@\S+", line))

print(list_emails) 
```

```python
#Zadatak 2

import re
list_atleastOneA=[]
list_onlyOneA=[]
list_noA=[]
list_numbers=[]
list_onlyNonCapitalLetters=[]


fhand = open('mbox-short.txt') 
for line in fhand:
  line = line.rstrip()
  if(re.findall("\S+a+\S+@\S+", line) != []):
    list_atleastOneA.append(re.findall("(\S+a+\S+)@\S+", line))
  if(re.findall("[^a\s]*a[^a\s]*@\S+", line) != []):
    list_onlyOneA.append(re.findall("([^a\s]*a[^a\s]*)@\S+", line))
  if(re.findall("[^a\s]+@\S+", line) != []):
    list_noA.append(re.findall("([^a\s]+)@\S+", line))
  if(re.findall("[^0-9\s]*[0-9]+[^0-9\s]*@\S+", line) != []):
    list_numbers.append(re.findall("([^0-9\s]*[0-9]+[^0-9\s]*)@\S+", line))
  if(re.findall("[a-z]+@\S+", line) != []):
    list_onlyNonCapitalLetters.append(re.findall("([a-z]+)@\S+", line))   

print(list_atleastOneA) 
print(list_onlyOneA) 
print(list_noA) 
print(list_numbers) 
print(list_onlyNonCapitalLetters) 
```
